### 1° Task

The Gaussian function can be written as: | La función gaussiana centrada en cero puede ser escrita como:

$$ g(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{ -\frac{1}{2}\left(\frac{x}{\sigma}\right)^2 } $$

Let's analyze how different values of sigma modify the shape of the $g(x)$ curve using plots. | Analicemos gráficamente cómo diferentes valores de sigma modifican la forma de la curva de $g(x)$.


1. Graph the Gaussian function with at least three different values of $\sigma$. | Realice una gráfica de la función gaussiana con al menos tres valores de $\sigma$.
2. The plots must have: | La gráfica debe tener:
     - both axes identified, | ambos ejes identificados
     - a title | un título
     - a legend | una leyenda

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-25,25,1)
g1 = 1
g2 = 4
g3 = 15

y1= 1/(np.sqrt(2.*np.pi)*g1)*np.exp(-((x/g1)**2)/2)
y2= 1/(np.sqrt(2.*np.pi)*g2)*np.exp(-((x/g2)**2)/2)
y3= 1/(np.sqrt(2.*np.pi)*g3)*np.exp(-((x/g3)**2)/2)
```

```python
plt.plot (x, y1, linestyle="-", label= "Gamma = 1")
plt.plot(x, y2, linestyle="-.", label= "Gamma = 4")
plt.plot (x, y3, linestyle="--", label= "Gamma = 15")
plt.legend()
plt.ylabel('Frecuencia')
plt.xlabel('Variable')
plt.title('Exploración del valor Gamma')
plt.show()
```
